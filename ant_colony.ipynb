```
import acopy
solver = acopy.Solver(rho=.03, q=1)
colony = acopy.Colony(alpha=1, beta=3)

#timer = acopy.plugins.TimerPlugin() # time runs
elite = acopy.plugins.EliteTracer(factor=2) # traces the elite ant and gives a pheromone bonus factor
threshold = acopy.plugins.Threshold(threshold=1719) # If a solution <= the threshold then the solver terminates early

plugins = [elite, threshold]
for p in plugins:
  solver.add_plugin(p)
```

```
tour = solver.solve(graph, colony, limit=100) #100 iterations
```
