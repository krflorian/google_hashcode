```
# setup
import os
import pandas as pd
from pulp import *

#os.chdir('google_hashcode/2020/practice')
os.getcwd()
```

read data

```
def read_data(filename):
  with open('data/'+filename, 'r', encoding='ascii', newline = '\n') as f:
    #data = [[val for val in line] for line in f]
    data =  f.readlines()
    data = [d.replace('\n', '').split() for d in data] 
  return data
```

```
filename = 'a_example.in'
data = read_data(filename)
data
```

setup classes


```
class pizza_problem():
  def __init__(self, max_slices):
    self.max_slices = max_slices
    self.pizzas = []

class pizza():
  def __init__(self, n_slices):
    self.slices = n_slices
```

```
problem = pizza_problem(int(data[0][0]))

for i in range(int(data[0][1])):
  problem.pizzas.append(pizza(int(data[1][i])))
problem.pizzas
```

```
problem.pizzas[0].slices
```

```
# Setup Milp

prob = LpProblem("The Pizza Problem", LpMaximize)

# create decision variables
decision_vars = LpVariable.dicts('pizzas', problem.pizzas, cat='Binary')
# create objective function
prob += lpSum([decision_vars[pizza]*pizza.slices for pizza in problem.pizzas]), "Total Pizza Slices"
```

```
# Constraints

prob += lpSum([decision_vars[pizza]*pizza.slices for pizza in problem.pizzas]) <= problem.max_slices, 'max slices'

prob
```

```
prob.solve()
```

```
pizzas_ordered = []

for pizza in problem.pizzas:
  if decision_vars[pizza].varValue > 0:
        pizzas_ordered.append(pizza)
pizzas_ordered
```

```
print('maximum pizzas: {} \npizzas ordered: {}'.format(problem.max_slices, sum([p.slices for p in pizzas_ordered])))
```
